## Recrified Linear Activation Function



```python
import numpy as np

weight1= -1
bias1= 0.5
weight2= -2
bias2= 1

inputs= np.linspace(-1, 1, 100)
outputs= []

import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x) 

for input_ in inputs:
    output1= relu(input_ * weight1 + bias1)
    output2= relu(output1 * weight2 + bias2)
    outputs.append(output2)

plt.plot(inputs, outputs)
plt.ylim(-2, 2)
plt.grid()
plt.show()
```

### What are axis and keepdims?

```python
a= [[1, 2, 3],
    [100, 200, 300], 
    [1000, 2000, 3000]]

# axis 0 = get sum per column
print(np.sum(a, axis= 0))
print(np.sum(a, axis= 0, keepdims= True))

print('\n')

# axis 1 = get sum per row
print(np.sum(a, axis= 1))
print(np.sum(a, axis= 1, keepdims= True))
```

```python
layer_outputs = np.array([[4.8, 1.21, 2.385], # sample 1 neuron 123
                          [8.9, -1.81, 0.2], # sample 2 neuron 123
                          [1.41, 1.051, 0.026]]) # sample 3 neuron 123

exp_values= np.exp(layer_outputs)
print(f'exp values :\n{exp_values}')

sum_exp_values= np.sum(exp_values, axis= 1, keepdims= True)
print(f'\nsum exp values :\n{sum_exp_values}')

norm_exp_values= exp_values / sum_exp_values
print(f'\nnormalized exp values :\n{norm_exp_values}')

sum_norm_exp_values= np.sum(norm_exp_values, axis= 1, keepdims= True)
print(f'\nsum of normalized exp values :\n{sum_norm_exp_values}')
```

```python
class ActivationSoftmax:

    def forward(self, inputs):

        # get unnormalized proba
        # subtract each number by max to guard from value explosion
        exp_values= np.exp(inputs - np.max(inputs, axis= 1, keepdims= True))

        # normalize exp_values to get probability
        proba= exp_values / np.sum(exp_values, axis= 1, keepdims= True)

        self.output= proba
```

### Why should include `-np.max(...)` in `np.exp`?

```python
inputs = [1, 100, 1000]

# Taking exponent of big numbers will cause value explosion.
np.exp(inputs)
```

```python
inputs= [1, 5, 10]

# But subtracting them by max values will not change exp
# since normalized exp care about the absolute distance, not the scale
# this will guard from value explosion

print(np.exp(inputs) / np.sum(np.exp(inputs)))
print(np.exp(inputs - np.max(inputs)) / np.sum(np.exp(inputs - np.max(inputs)) ))
```

```python
softmax = ActivationSoftmax()

inputs= np.array([[1, 2, 3]])
softmax.forward(inputs)
softmax.output
```

```python
# 'what happen if we subtract 3 from all of them?

inputs= np.array([[-2, -1, 0]])
softmax.forward(inputs)
softmax.output
```
