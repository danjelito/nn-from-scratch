To explain the order of parameters we are passing into ​np.dot(),​ we should think of it as whatever
comes first will decide the output shape.

```python
a = [1, 2, 3] 
b = [[2, 3, 4], 
     [5, 6, 7]]

np.dot(b, a)
```

We get dot product of: 
1. b[0] with a 
2. b[1] with a

To perform a matrix product, the size of the second dimension of the left matrix must match the
size of the first dimension of the right matrix. 

For example, if the left matrix has a shape of ​(5, 4)
then the right matrix must match this 4 within the first shape value ​(4, 7)​. 

The shape of the
resulting array is always the first dimension of the left array and the second dimension of the right
array, ​(5, 7).

In the case like above, we must transpose `weights` so that it becomes a (4, 3) matrix.
