# Linear Function

```python
def f(x):
    return 2*x

X= np.array(range(5))
y= f(X)

X, y
```

```python
plt.plot(X, y)
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.show()
```

slope= change_in_y / change_in_x 

slope= delta y / delta x

```python
slope= (y[1] - y[0]) / (X[1] - X[0])
slope
```

We can say that everytime X increase by 1, y increase by 2.

# Nonlinear Function

```python
def f(x):
    return 2*x**2

X= np.array(range(5))
y= f(X)

print(X, y)

plt.plot(X, y)
plt.xlim(0, 5)
plt.ylim(0, 50)
plt.show()
```

```python
slope= (y[1] - y[0]) / (X[1] - X[0])
slope
```

```python
slope= (y[2] - y[1]) / (X[2] - X[1])
slope
```

For nonlinear function, the slope keeps changing.

Derivative = slope at one point. To measure this, we use two points which are infinitely close (with limit).

```python
# slope at x = 1

delta= 1e-4
x1= 1
x2= x1 + delta

y1= f(x1)
y2= f(x2)

slope= (y2 - y1) / (x2 - x1)
slope
```
